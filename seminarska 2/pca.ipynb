```python
# imports

import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

## Word2Vec principle component analysis

```python
# Load the dataset
data = pd.read_csv("w2v.csv")

# Separate the features from the target variable
y = data["category"]
X = data.drop("category", axis=1)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a new DataFrame with the PCA results
pca_df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
pca_df["category"] = y

# Print the first few rows of the PCA DataFrame
print(pca_df.head())

# Plot the PCA results
sns.scatterplot(x="PC1", y="PC2", hue="category", data=pca_df, s=2)
plt.show()
```

## SBERT embedding principle component analysis

```python
# Load the dataset
emb_df = pd.read_csv("embeddingsCTXencode_plus_Sbert.csv", header=None) # embeddings
data = pd.read_json("dataset.json") # targets

# Separate the features from the target variable
y = data["category"]
X = emb_df

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a new DataFrame with the PCA results
pca_df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
pca_df["category"] = y

# Print the first few rows of the PCA DataFrame
print(pca_df.head())

# Plot the PCA results
sns.scatterplot(x="PC1", y="PC2", hue="category", data=pca_df, s=2)
plt.show()
```
